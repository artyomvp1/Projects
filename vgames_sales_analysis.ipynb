```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
df = pd.read_csv('vgsales.csv')
df.sort_values('NA_Sales', ascending=False).head(3)
```

## <font color='crimson'>1. Peak of video games</font>

```python
bins = np.arange(df['Year'].min(), 2020, 1)

plt.hist(df['Year'], bins=bins, color='#18a89c')

plt.title('Number of Games Sold', fontdict={'fontweight':'bold', 'fontsize':20}, loc='right')
plt.xlabel('Year', fontsize=16)
plt.ylabel('Number of games released', fontsize=14)

# plt.savefig('g1.png', dpi=300)
plt.show()
```

## <font color='crimson'>2. Genre</font>

```python
genre = [i for i in df['Genre'].unique()]
weights = [df.loc[df['Genre'] == i].count()[0] for i in genre]
# explode = [0.08, 0, 0, 0, 0, 0, 0, 0, 0.12, 0, 0, 0]
explode = [0.12 if i == 'Sports' else 0.15 if i == 'Action' else 0 for i in genre] # elif
```

```python
plt.figure(figsize=(7,7))

plt.pie(weights, labels=genre, autopct='%.1f %%', pctdistance=0.8, explode=explode)

# plt.savefig('g2.png', dpi=300)
plt.show()
```
